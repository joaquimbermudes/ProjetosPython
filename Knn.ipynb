# K-Nearest-Neighbor (knn)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Começamos criando três variáveis relacionadas de modo não-linear para mostrar o poder de adaptabilidade do knn.

A variável $x$ será distribuida uniformomente entre no intervalo $[0,1]$ com $200$ observações. Uma variável de ruído chamada de "epsilon" ($ϵ$) terá distribuição $N(0, 100)$.  A variável $y$ obedecerá a relação 
$$y = x - 3x\cos(x) + 5 + 10 \sin(x) + ϵ$$

```python
x = np.random.uniform(0, 20, 200)
epsilon = np.random.normal(0, 10, 200)

def f(x):
  return x - 3*x*np.cos(x) + 5 + 10*np.sin(x)

y =  f(x) + epsilon

plt.scatter(x, y)
```

Agora vamos para o knn propriamente dito.

O knn é um algorítmo de aprendizado supervisionado. Ele recebe o input de um elemnto $x_i$ e então procura em toda base de dados quais são os $k$ elementos mais próximos de $x_i$ e calcula a média. Esse método pode ser modificado alterando o parâmetro $k$, o modo de medir a distância e a função calculada com os $k$ visinhos mais proximos.

```python
def Knn(x0, y0, k):
  z = x0
  ind = range(len(x0))
  estimado = np.zeros(len(y0))
  for i in range(len(x0)):
    z = np.abs(x0 - x0[i])
    ind = np.argsort(z)
    estimado[i] = np.mean(y0[ind][1:k])
  return estimado
```

Um exemplo de aplicação de knn usando os 10 vizinhos mais próximos.

```python
plt.scatter(x, y)
plt.scatter(x, Knn(x0 = x, y0 = y, k = 10))
plt.plot(np.arange(0, 20, 0.05), f(np.arange(0, 20, 0.05)), color = "red")
```

Um exemplo de aplicação para os 2 vizinhos mais próximos.

```python
plt.scatter(x, y)
plt.scatter(x, Knn(x0 = x, y0 = y, k = 2))
plt.plot(np.arange(0, 20, 0.05), f(np.arange(0, 20, 0.05)), color = "red")
```
