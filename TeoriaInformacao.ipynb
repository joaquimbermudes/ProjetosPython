**DEFINIÇÂO:** A entropia H(x) de uma variável aleatória discreta X é definida da seguinte maneira

$$
H(X) = - \sum_{x \in X}{p(x)}log_{2}p(x)
$$

A entropia H(x) é dado por um número real maior ou igual à zero. Esse número é aproximadamente a esperança da quantidade de perguntas binárias para decobrir o valor de x do resultado mais provável até o menos provável. 

Ex:
    P(x = 0) = 0.05
    P(x = 1) = 0.5
    P(x = 2) = 0.3
    P(x = 3) = 0.15

A esperança do número de perguntas é cotado por baixo pela entropia H(x) e por cima pela entropia H(x) + 1.

```python
rep = 1000
conjunto = 4
prob = [0.05, 0.5, 0.3, 0.15]

escolha = np.random.choice(conjunto, rep, prob)

perguntas = np.zeros(rep)

for i in range(rep):
    if escolha[i] == 1:
        perguntas[i] = 1
    elif escolha[i] == 2:
        perguntas[i] = 2
    else:
        perguntas[i] = 3


print("H =",sum(prob*(-np.log2(prob))), " < Perguntas =", np.mean(perguntas), "< H + 1 =",sum(prob*(-np.log2(prob)))+1)
```

A entropia H(x) é convexa em P e atinge valor máximo no caso da distribuição uniforme.

```python
prob = np.arange(1, 100)/100
prob2 = np.ones(99) - prob

H = - (prob * np.log2(prob) + prob2 * np.log2(prob2))

plt.plot(prob, H)
```

**DEFINIÇÂO:** A entropia relativa D(p||q) é dado por

$$
D(p||q) = - \sum_{x \in \Chi}{p(x)}log_{2} \frac{p(x)}{q(x)}
$$

D(p||q) é o aumento da entropia que acontece quando é considerado uma distribuição de probabilidade errada.

Ex:
    P(x = 0) = 0.7
    P(x = 1) = 0.3

Na figura abaixo podemos ver que D(p||q) é minimizado quado consideramos a probabilidade correta.

```python
p = 0.7
prob_real = [p, 1 - p]

prob = np.arange(1, 100)/100
prob2 = np.ones(99) - prob


D = prob_real[0] * np.log2(prob_real[0]/prob) + prob_real[1] * np.log2(prob_real[1]/prob2)

plt.plot(prob, D)
```

$$

$$

I(X;Y) é o aumento na entropia que acontece quando consideramos que duas variáveis aleatórias são independentes.
