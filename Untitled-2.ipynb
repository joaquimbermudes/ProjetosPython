```python
import numpy as np
import seaborn as sns
import imageio
import matplotlib.pyplot as plt
import sympy as sp
```

Gerando uma matriz de variáveis aleatórias de 500 linhas e 5000 colunas. Aqui usei uma distribuição expondencial, mas pode ser qualquer uma.

```python
N = (500, 5000)
dist_exp = np.random.exponential(10, N)
```

***TEOREMA CENTRAL DO LIMITE:*** Sejam $X_{1} ... X_{n}$ uma amostra IID com média $\mu$ e variância $\sigma^{2} < \infty$. Definimos $\bar{X}_{n} = n^{-1} \sum_{i=1}^{n}X_{i}$, então:

$$
Z_{n} \equiv \frac{\sqrt{n}(\bar{X}_{n} - \mu)}{\sigma} \rightarrow Z
$$

$Z_n$ converge em distribuição para $Z \sim N(0,1)$.

```python
n = np.arange(1,N[0]+1)
X = np.zeros(N[1])
for i in range(0,N[1]):
    X[i] = np.mean(dist_exp[:, i])
tcl = (X-10)*np.sqrt(500)/10
sns.displot(tcl)
print(np.std(tcl), np.mean(tcl))
```

```python
np.random.seed(50)

gif_path = "test.gif"
frames_path = "Figuras_para_gif/{i}.jpg"

for i in range(1, 60):
    F = 500

    distX = np.random.exponential(3, F)
    distC = np.random.normal(5, 10/i, F)
    
    figura = sns.displot(distX + distC, kind = "hist", bins = 50)
    figura.set_axis_labels("Distribuição expondencial com média 3 + distribuição normal com média 5 e desvio padrão 10/{i}".format(i=i))
    plt.xlim(-8, 30)
    plt.ylim(0, 100)


    figura.savefig("Figuras_para_gif/{i}.jpg".format(i=i))

with imageio.get_writer(gif_path, mode='I') as writer:
    for i in range(1,60):
        writer.append_data(imageio.imread(frames_path.format(i=i)))
```
